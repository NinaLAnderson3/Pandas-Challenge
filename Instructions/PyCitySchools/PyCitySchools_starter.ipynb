```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculuate number of unique school names
grouped_school = school_data_complete.groupby(['school_name']).count()
num_school=grouped_school["School ID"].count()
num_school
```

```python
#calculate number of total students by "Student ID"
num_stu = grouped_school.sum()["Student ID"]
num_stu
```

```python
#sum up unique budgets
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
#calculate average reading score
avg_reading = school_data_complete["reading_score"].mean()
avg_reading
```

```python
#calculate average math score
avg_math = school_data_complete["math_score"].mean()
avg_math
```

```python
#calculate number students with math score greater than or equal to 70
num_pass_math = school_data_complete.loc[school_data_complete["math_score"]>=70].count()["Student ID"]
num_pass_math
```

```python
#calculate number students with reading score greater than or equal to 70
num_pass_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70].count()["Student ID"]
num_pass_reading
```

```python
#calculate number of students with math AND reading score greater than or equal to 70
passing_both = school_data_complete.loc[(school_data_complete["math_score"]>=70) & 
                                        (school_data_complete["reading_score"]>=70)].count()["Student ID"]
passing_both
```

```python
#caluclate percent of students passing math
perc_pass_math = (num_pass_math/num_stu)*100
perc_pass_math
```

```python
#calculate percent of students passing reading
perc_pass_reading = (num_pass_reading/num_stu)*100
perc_pass_reading
```

```python
#calculate percent of students passing both math and reading
perc_pass_both = (passing_both/num_stu)*100
perc_pass_both
```

```python
#create new dataframe

dist_summary = pd.DataFrame({"Total Number of Schools":[num_school],
                            "Total Number of Students":[num_stu],
                            "Total Budget":[total_budget],
                            "Average Reading Score":[avg_reading],
                            "Average Math Score":[avg_math],
                            "Percent Passing Reading":[perc_pass_reading],
                            "Percent Passing Math":[perc_pass_math],
                            "Percent Passing Both":[perc_pass_both]})
#format dataframe
dist_summary["Total Budget"] = dist_summary["Total Budget"].map("${:,.2f}".format)
dist_summary["Average Reading Score"] = dist_summary["Average Reading Score"].map("{:,.2f}%".format)
dist_summary["Average Math Score"] = dist_summary["Average Math Score"].map("{:,.2f}%".format)
dist_summary["Percent Passing Math"] = dist_summary["Percent Passing Math"].map("{:,.2f}%".format)
dist_summary["Percent Passing Reading"] = dist_summary["Percent Passing Reading"].map("{:,.2f}%".format)
dist_summary["Percent Passing Both"] = dist_summary["Percent Passing Both"].map("{:,.2f}%".format)
dist_summary
```

```python
#create table with school name, school type, total students, total budget
first_df = school_data_complete[["school_name","type","budget"]].drop_duplicates()
first_df
```

```python
stu_per_school = school_data_complete.groupby('school_name').count()["Student ID"]
stu_per_school
```

```python
#calculatae average math scores by school
avg_math = school_data_complete.groupby('school_name').mean()["math_score"]
avg_math
```

```python
#calculatae average reading scores by school
avg_reading = school_data_complete.groupby('school_name').mean()["reading_score"]
avg_reading
```

```python
#Find the number of students who passed BOTH reading AND math at each school, then divide by total # stu at school
pass_both = school_data_complete.loc[school_data_complete["reading_score"] & school_data_complete['math_score']>=70]
pass_both_grouped = pass_both.groupby('school_name').count()["Student ID"]
pass_both_grouped
perc_pass_both = (pass_both_grouped/stu_per_school)*100
perc_pass_both
```

```python
#put calculated scores and passing percentages into datafram
scores_df = pd.DataFrame({"Average Math Score":avg_math,
                     "Percent Passing Math":perc_pass_math,
                         "Average Reading Score": avg_reading,
                         "Percent Passing Reading":perc_pass_reading,
                         "Percent Passing Both":perc_pass_both})
#format dataframe
scores_df["Average Math Score"] = scores_df["Average Math Score"].map("{:,.2f}%".format)
scores_df["Average Reading Score"] = scores_df["Average Reading Score"].map("{:,.2f}%".format)
scores_df["Percent Passing Math"] = scores_df["Percent Passing Math"].map("{:,.2f}%".format)
scores_df["Percent Passing Reading"] = scores_df["Percent Passing Reading"].map("{:,.2f}%".format)
scores_df["Percent Passing Both"] = scores_df["Percent Passing Both"].map("{:,.2f}%".format)
scores_df
```

```python
budget_df = pd.merge(first_df,stu_per_school,on = 'school_name',how = 'left')
budget_df = budget_df.rename(columns={"Student ID":"Number of Students"})
budget_df["Per Student Budget"] = budget_df['budget']/budget_df["Number of Students"].astype(float)
budget_df
```

```python
#calculate number of students at each school that have reading grade >=70
pass_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70]
pr_grouped = pass_reading.groupby(['school_name']).count()["Student ID"]
pr_grouped
#calculate percent of students passing reading at each school by diviging # passing reading by totla # stu at school
perc_pass_reading = (pr_grouped/stu_per_school)*100
perc_pass_reading

scores_df = pd.DataFrame({"Average Reading Score":avg_reading,
                          "Percent Pass Reading":perc_pass_reading,
                          "Average Math Score": avg_math,
                          "Percent Pass Math":perc_pass_math})

scores_df
```
