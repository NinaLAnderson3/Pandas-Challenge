```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

school_data_complete.head()

```python
# grouping school data to count number of school IDs
grouped_school = school_data_complete.groupby(['school_name']).count()
num_school=grouped_school.count()["School ID"]
num_school
```

```python
#calculate total number of students based on unique Student ID #s
num_stu = grouped_school.sum()["Student ID"]
num_stu
```

```python
#calculate total budget 
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
#calculate average reading scores
avg_reading = school_data_complete["reading_score"].mean()
avg_reading
```

```python
#calculate average math scores
avg_math = school_data_complete["math_score"].mean()
avg_math
```

```python
#find the number of students passing math 
num_pass_math = school_data_complete.loc[school_data_complete["math_score"]>=70].count()["Student ID"]
num_pass_math
```

```python
#find the number of students passing reading
num_pass_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70].count()["Student ID"]
num_pass_reading
```

```python
#calculate number of students passing both
passing_both = school_data_complete.loc[(school_data_complete["math_score"]>=70) & 
                                        (school_data_complete["reading_score"]>=70)].count()["Student ID"]
passing_both
```

```python
#calculate percent passing math
perc_pass_math = (num_pass_math/num_stu)*100
```

```python
#calculate percent passing reading
perc_pass_reading = (num_pass_reading/num_stu)*100
```

```python
#calculate percent passing both
perc_pass_both = (passing_both/num_stu)*100
perc_pass_both
```

```python
#put calculations into a dataframe
dist_summary = pd.DataFrame({"Total Number of Schools":[num_school],
                            "Total Number of Students":[num_stu],
                            "Total Budget":[total_budget],
                            "Average Reading Score":[avg_reading],
                            "Average Math Score":[avg_math],
                            "Percent Passing Reading":[perc_pass_reading],
                            "Percent Passing Math":[perc_pass_math],
                            "Percent Passing Both":[perc_pass_both]})
#format columns in the dataframe
dist_summary["Total Budget"] = dist_summary["Total Budget"].map("${:,.2f}".format)
dist_summary["Average Reading Score"] = dist_summary["Average Reading Score"].map("{:,.2f}%".format)
dist_summary["Average Math Score"] = dist_summary["Average Math Score"].map("{:,.2f}%".format)
dist_summary["Percent Passing Math"] = dist_summary["Percent Passing Math"].map("{:,.2f}%".format)
dist_summary["Percent Passing Reading"] = dist_summary["Percent Passing Reading"].map("{:,.2f}%".format)
dist_summary["Percent Passing Both"] = dist_summary["Percent Passing Both"].map("{:,.2f}%".format)
dist_summary
```
